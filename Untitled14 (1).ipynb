```python
from google.colab.patches import cv2_imshow
import cv2 as cv
import numpy as np

# Load the query image in grayscale (for SIFT)
img1 = cv.imread('/content/scene2-scaled.jpg', cv.IMREAD_GRAYSCALE)

# Load the target image in color for final drawing
img2_color = cv.imread('/content/scene1-scaledlight.jpg')  # Load in color
img2_gray = cv.cvtColor(img2_color, cv.COLOR_BGR2GRAY)     # Convert to grayscale for SIFT

# Initialize SIFT detector
sift = cv.SIFT_create()

# Find keypoints and descriptors
kp1, des1 = sift.detectAndCompute(img1, None)
kp2, des2 = sift.detectAndCompute(img2_gray, None)

# FLANN matcher parameters
FLANN_INDEX_KDTREE = 1
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
search_params = dict(checks=50)

flann = cv.FlannBasedMatcher(index_params, search_params)
matches = flann.knnMatch(des1, des2, k=2)

# Lowe's ratio test
good = []
for m, n in matches:
    if m.distance < 0.7 * n.distance:
        good.append(m)

# Minimum number of matches
MIN_MATCH_COUNT = 10

if len(good) > MIN_MATCH_COUNT:
    src_pts = np.float32([kp1[m.queryIdx].pt for m in good]).reshape(-1, 1, 2)
    dst_pts = np.float32([kp2[m.trainIdx].pt for m in good]).reshape(-1, 1, 2)

    # Find homography
    M, mask = cv.findHomography(src_pts, dst_pts, cv.RANSAC, 5.0)

    # Draw bounding box
    h, w = img1.shape
    pts = np.float32([[0, 0], [0, h - 1], [w - 1, h - 1], [w - 1, 0]]).reshape(-1, 1, 2)
    dst = cv.perspectiveTransform(pts, M)

    # Draw box on color image
    img2_marked = cv.polylines(img2_color.copy(), [np.int32(dst)], True, (0, 255, 0), 3, cv.LINE_AA)

    # Show the image with the detected object
    cv2_imshow(img2_marked)
else:
    print(f"Not enough matches are found - {len(good)}/{MIN_MATCH_COUNT}")
```
